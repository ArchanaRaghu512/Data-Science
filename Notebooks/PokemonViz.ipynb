```python
type2_df = data["Type 2"].value_counts()
#sns.countplot(data=data,"Type 2")
%matplotlib inline
sns.countplot('Type 2',data=data).set_title("Representation of Null values in Type 2")
fig = plt.gcf()
fig.set_size_inches(18,8)
plt.show()
```

```python
sns.lmplot(x="HP",y="Speed", data=data, hue= "Type 2",fit_reg=None)

sns.plt.title("HP Vs Speed")
fig=plt.gcf()
fig.set_size_inches(15,9)
plt.show()
```

```python
box_df = data.drop(["#","Generation","Legendary"], axis=1)
sns.boxplot(data=box_df)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```

```python
fig=plt.gcf()
fig.set_size_inches(15,8)
sns.violinplot(x="Type 1",y="Sp. Atk", data=data, inner=None)
sns.set_style("whitegrid")
sns.plt.title("Type1 Vs Special Attack")
```

```python
fig=plt.gcf()
fig.set_size_inches(15,8)
sns.swarmplot(x="Type 1",y="Sp. Atk", data=data, color="k",alpha=0.5)
sns.violinplot(x="Type 1",y="Sp. Atk", data=data, inner =None)
```

```python
fig=plt.gcf()
fig.set_size_inches(15,8)
sns.distplot(data.Speed)
```

```python
corr = data.corr()
corr
```

```python
fig=plt.gcf()
fig.set_size_inches(15,12)
sns.plt.title("Correlation Plot")
sns.heatmap(corr, annot=True, linewidths=0.3)
```

```python
fig=plt.gcf()
fig.set_size_inches(19,15)
sns.jointplot(x="Defense",y="Attack", data=data)
plt.show()
```
